5.SELECT model,rating FROM campusx.smartphones

1. SELECT brand_name,count(*) as 'count',round(avg(rating)) as 'avg_rating' FROM campusx.smartphones

2.SELECT brand_name,avg(ram_capacity) as 'avg_ram' FROM campusx.smartphones
