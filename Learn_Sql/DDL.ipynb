### constraints in SQL

##### 1. not null

    user_id integer not null, --> it will never null

    password varchar(255)

##### 2. unique

    name varchar(255) not null,

    password varchar(255) not null

##### -->. second wa to create a constraints

#### 3. primary key

create table users (
    user_id integer not null primary key,
    name varchar(255) not null,
    email varchar(255) not null,
    password varchar(255) not null,
    ### another way
    constraint users_pk primary_key (user_id)

)


### 4. auto increment :- isse auto inctement hota hai jis-bhi column pr aap lagate ho

    user_id integer not null primary key auto_increment, 
    name varchar(255) not null,
    email varchar(255) not null,
    password varchar(255) not null,

)

##### 5. check :- condition apply krta hai

##### 6. Default constraint

create table ticket (
    ticket_id integer primary key,
    name varchar(255) not null,
    travel_date timestamp default current_data
)


### 7. foreign key -> jb ak table ka primary key is available in another table to establish a relationship b/w two table that is called foreign key


create table customers (
    cid integer primary key auto_increment,
    name varchar(255) not null,
    email varchar(255) not null unique
)

### now we are making another (order) table

create table orders (
    order_id integer primary key auto_increment,
    cid integer not null,
    order_date datetime not null default current_timestamp,
    contraint orders_fk foreign key (cid) references customers(cid)
)
### if you want to delete some rows from (customer) table then you can't.

1. foreign key aese smjho ki aap kisi customer ko delete krne ja rhe ho aur uska order available hai to wo delete nhi hoga kuki uska order available hai order table me

2. isiliye data integrity ko mentain krne keliye hmlog foreign key ka use kr skte hai



#### referential-actions -> agr 2 table ak dusre se related hai using foreign key to kisi ak table me update ,delete krne me how 2d table will respond

1. restrict = it will not allow to delete

2. cascade = iska matlab ye hai ki aap agr customer table me kuch bhi change kijiyega to wo order me bhi change ho jayega

3.set-null = isko krne se agr kuch delete hota hai to waha pr "Null" aa jata hai

4. set-default = ye jo default value set krenge whi delete hone ke baad aa jayega
