```python
# Brute force without chr()

# build manual ASCII-to-character mapping for AZ and az
ascii_map = {}
num = 65
for ch in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":
    ascii_map[num] = ch
    num += 1

num = 97
for ch in "abcdefghijklmnopqrstuvwxyz":
    ascii_map[num] = ch
    num += 1

# take four inputs separately
nums = list(map(int,input("Enter numbers").split()))
for numss in nums:
    print(f"{numss} - {ascii_map[numss]}")
```

```python
# create mapping for uppercase and lowercase letters
ascii_map = {ord(c): c for c in "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz"}

# read all four numbers in one line
nums = list(map(int, input("Enter 4 digits: ").split()))

for n in nums:
    if n in ascii_map:
        print(f"{n}-{ascii_map[n]}")
    else:
        print(f"{n}-?")
```
