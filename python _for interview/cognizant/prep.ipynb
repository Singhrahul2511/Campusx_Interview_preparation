```python
# optimal approch

num = int(input("Enter any number: "))
if num<0:
    print("Not valid number")
else:
    temp = num
    rev_num = 0
    while num>0:
        rem = num%10
        rev_num = rev_num*10 +rem
        num //=10
    if rev_num == temp:
        print("Palindrome")
    else:
        print("Not a Plindrome")
```

```python
### count number f digits in an number
num = int(input("Enter any number: "))
cnt = 0             # 0(log(n))
while num>0:
    num //=10
    cnt +=1
print(cnt)
```

```python
### Armstrong number
num = int(input("Enter any number: "))
temp =num
temp1 = num
cnt = 0
while num>0:
    num //=10
    cnt +=1
sum = 0
while temp>0:
    lst_dig = temp%10
    sum +=lst_dig**cnt
    temp //=10
if sum == temp1:
    print("arms")
else:
    print("not")



```

```python
## print all factors of any number in list

num = int(input("Enter any number: "))
lst = []
for i in range(1,num+1):
    if num%i==0:
        lst.append(i)
print(lst)
```

```python
## calculate factorial of an number

from math import sqrt
num = int(input("Enter any number: "))
res = []
for i in range(1,int(sqrt(num))+1):
    if num%i ==0:
        res.append(i)           # tc = 0(sqrt(n)) + 0(nlogn)
        if num//i !=i:
            res.append(num//i)
res.sort()
print(res)
```
